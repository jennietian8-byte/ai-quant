## 指标计算公式

- RSI(14)：涨跌幅 -> 平均上涨/平均下跌 -> RS -> RSI。
- MACD(12,26,9)：EMA12、EMA26、DIF、DEA、MACD柱。
- 布林带(20,2)：MA20、上轨、下轨、带宽。
- ATR(14)：TR 三项最大值的 14 日移动平均；ATR_PCT = ATR / close × 100%。
